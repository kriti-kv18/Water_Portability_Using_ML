```python
import matplotlib.pyplot as plt

classifiers = ['Logistic Regression', 'Naive Bayes', 'Extra Trees Classifier', 'Random Forest', 'KNN', 'Decision Tree Classifier', 'Ridge Classifier']

# Sample scores for F1 score, recall, precision, Kappa, and MCC
f1_scores = [0.55, 0.60, 0.85, 0.82, 0.60, 0.80, 0.55]
recall_scores = [0.60, 0.65, 0.80, 0.78, 0.65, 0.75, 0.60]
precision_scores = [0.50, 0.55, 0.90, 0.88, 0.55, 0.85, 0.50]
kappa_scores = [0.50, 0.55, 0.85, 0.82, 0.55, 0.80, 0.50]
mcc_scores = [0.45, 0.50, 0.80, 0.78, 0.50, 0.75, 0.45]

# Plot F1 score
plt.figure(figsize=(10, 6))
plt.bar(classifiers, f1_scores, color='lightgreen')
plt.title('Comparison of Classifier F1 Scores for Water Quality Analysis')
plt.xlabel('Classifiers')
plt.ylabel('F1 Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot recall
plt.figure(figsize=(10, 6))
plt.bar(classifiers, recall_scores, color='coral')
plt.title('Comparison of Classifier Recall Scores for Water Quality Analysis')
plt.xlabel('Classifiers')
plt.ylabel('Recall')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot precision
plt.figure(figsize=(10, 6))
plt.bar(classifiers, precision_scores, color='lightblue')
plt.title('Comparison of Classifier Precision Scores for Water Quality Analysis')
plt.xlabel('Classifiers')
plt.ylabel('Precision')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot Kappa
plt.figure(figsize=(10, 6))
plt.bar(classifiers, kappa_scores, color='lightsalmon')
plt.title('Comparison of Classifier Kappa Scores for Water Quality Analysis')
plt.xlabel('Classifiers')
plt.ylabel('Kappa')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot MCC
plt.figure(figsize=(10, 6))
plt.bar(classifiers, mcc_scores, color='yellow')
plt.title('Comparison of Classifier MCC Scores for Water Quality Analysis')
plt.xlabel('Classifiers')
plt.ylabel('MCC')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
X = data.drop("Potability", axis=1)
y = data["Potability"]

# Train ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X, y)

# Define predict function
def predict_water_potability(pH, Hardness, Solids, Chloramines, Sulfate, Conductivity, Organic_carbon, Trihalomethanes, Turbidity):
    # Predict water potability using the trained model
    input_data = [[pH, Hardness, Solids, Chloramines, Sulfate, Conductivity, Organic_carbon, Trihalomethanes, Turbidity]]
    result = model.predict(input_data)

    return 'Safe' if result == 1 else 'Not safe'

iface = gr.Interface(fn=predict_water_potability,
                     inputs=["number", "number", "number", "number", "number", "number", "number", "number", "number"],
                     outputs="text",
                     title="Water Potability Prediction",
                     description="Enter the water quality parameters to predict if it's safe for drinking.")

iface.launch()
```
